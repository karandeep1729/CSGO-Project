```python
# These 3 columns are categorical and these 3 columns are important to us.
# We need to convert these 3 columns map,round_winner and bomb_planted from
# categorical to numerical.

from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.8,random_state=29)
x_train
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# It is taking too much time to execute.
# Even passing 10000 as max_iter it gave us an warning. It means still the model is not
# able to cover the entire data.
```

```python
# Problem -> So Many Columns
# We came to the conclusion that the model is taking so much time to execute because of higher dimensions.
# Now we need to reduce the number of dimensions.

# To reduce the dimensions -> PCA and LDA
# Which is more suitable in this SCENARIO??
# # lda suitable for classification problem
# This is a supervised problem
# LDA works a lot better for classification tasks than PCA.
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
# Before applying LDA we need to scale our data. -> Standardization
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
lda.fit(x_train,y_train)
```

```python
lda.transform(x_test)
```

```python
pd.DataFrame(x_train)
```

```python
# How do we select the number of components??

# min(number_of_features , number_of_classes-1)
# min (96 , 2-1)
# min (96,1)
# Practically with respect to this data we cannot convert 96 columns into just one single column.
# Now we will see how we select the number of components manually by calculation lda_coeff.
```

```python
lda_coef = np.exp(abs(lda.coef_))
```

```python
lda_coef   # a 2d array.
```

```python
lda_coef.shape
```

```python
lda_coef= lda_coef.flatten()   # Used to convert any dimensions to 1d array
```

```python
lda_coef
```

```python
x
```

```python
feature_names = x.columns
feature_names
```

```python
df_feature_imp = pd.DataFrame({'Features':feature_names,'imp_value':lda_coef})
df_feature_imp
# Here we can see the importance of every feature/column.
# Higher the coeff means higher the importance.
```

```python
top_20 = df_feature_imp.sort_values(by='imp_value',ascending=False).head(20) # Randomly selecting 20.
```

```python
top_20
```

```python
imp_col = top_20.index
imp_col
```

```python
x_train = x_train[:,imp_col]
```

```python
pd.DataFrame(x_train)
```

```python
x_test = x_test[:,imp_col]
```

```python
x_test
```

```python
logistic_model = LogisticRegression()
```

```python
logistic_model.fit(x_train,y_train)
```

```python
y_pred = logistic_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# Before applying LDA -> 75% Accuracy
# Affer applying LDA -> 75 % Accuracy
# After LDA we reduced the number of dimensions from 96 -> 20 and still the model is working with same efficieny.
```
